## Ques 1(Amazon best selling books)


```python
import requests
from bs4 import BeautifulSoup
```

```python
def get_data(pageno):
  url='https://www.amazon.in/gp/bestsellers/books/ref=zg_bs_pg_'+str(pageno)+'?ie=UTF8&pg='+str(pageno)
  r=requests.get(url)
  soup=BeautifulSoup(r.content,'html.parser')#gives byte content(bytes),parser will parse through the tree
  complete=[]
  for d in soup.find_all('div',attrs={'class':"zg-grid-general-faceout"}):
    link=d.find('a',attrs={'class':"a-link-normal"})
    
    sub=[]
    if link is not None:
      sub.append('https://www.amazon.in/'+link.get('href'))#for accessing properties we use .get method
    else:
      sub.append('NaN')
    complete.append(sub)
  return complete
```

```python
number_of_pages = 2
result = []
for i in range(1, number_of_pages):
  result.append(get_data(i))
```

```python
import numpy as np
urls = np.array(result)
urls = urls.flatten()
urls
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import numpy as np
PATH="C:\Program Files (x86)\chromedriver.exe"

complete=[]

for url in urls:
    
    driver = webdriver.Chrome(PATH)
    driver.get(url)
    
    for books in driver.find_elements(By.ID,'detailBulletsWrapper_feature_div'):
        
        dict={
            "publisher":None,
            "language":None,
            "paperback":None,
            "ISBN-10":None,
            "ISBN-13":None,
            "weight":None,
            "dimensions":None,
            "reviews":None
            }     

        for t in books.find_elements(By.CLASS_NAME,"a-list-item"):
            
            split=t.text.split(' : ')
            if(split[0]=='Publisher'):
                dict['publisher']=split[1]
                
            if(split[0]=='Language'):
                dict['language']=split[1]
                                
            if(split[0]=='Paperback'):
                dict['paperback']=split[1]
                
            if(split[0]=='ISBN-10'):
                dict['ISBN-10']=split[1]
                
            if(split[0]=='ISBN-13'):
                dict['ISBN-13']=split[1]
                
            if(split[0]=='Item Weight'):
                dict['weight']=split[1]
                
            if(split[0]=='Dimensions'):
                dict['dimensions']=split[1]

            # top 3 reviews are collected below 
            i=1
            for reviews in driver.find_elements(By.ID,'cm-cr-dp-review-list'):
                rev_sub=[]
                for review in reviews.find_elements(By.CLASS_NAME,'review-text'):
                    
                    if i==3:
                        break
                    
                    if review.text is not None:
                        rev_sub.append(review.text)
                    else:
                        rev_sub.append(None)
                    i=i+1
                
                if rev_sub is not None:
                    dict['reviews']=rev_sub
                else:
                    dict["reviews"]=None
                
        complete.append(dict)  
    driver.quit()
```

## Ques 2(IMDB movies data collection)


```python
import numpy as np
import requests
from bs4 import BeautifulSoup

def IMDB_scraping(year,total_movies,url):
  url=url
  r=requests.get(url)
  soup=BeautifulSoup(r.content,'html.parser')

  complete=[]
  dur=[]
  movie_name=[]
  rating=[]
  desc=[]
  votes=[]
  meta_score=[]
  genre=[]
  directors_and_stars=[]
  dir=[]
  stars=[]
  
  for i in soup.find_all('div',attrs={'class':'lister-item-content'}):
    vote=i.find('span',attrs={'name':'nv'})
    meta=i.find('span',attrs={'class':'metascore favorable'})
    gen=i.find('span',attrs={'class':'genre'})
    duration=i.find('span',attrs={'class':'runtime'})
    
    # movie name
    for name_div in soup.find_all(class_='lister-item-header'):
      if name_div is not None:
        movie_name.append(name_div.find('a').text)
      else:
        movie_name.append(np.nan)

    # IMDB rating
    for rating_div in soup.find_all(class_='inline-block ratings-imdb-rating'):
      if rating_div is not None:
        rating.append(rating_div.find('strong').text)
      else:
        rating.append(np.nan)

    # description
    if (i.select("p:nth-of-type(2)")[0].text) is not None:
      desc.append(i.select("p:nth-of-type(2)")[0].text)
    else:
      desc.append(np.nan)
    
    # votes
    if vote is not None:
      votes.append(vote.text)
    else:
      votes.append(np.nan)
    
    directors_and_stars.append(i.select("p:nth-of-type(3)")[0].text)
    
    # metascore
    if meta is not None:
      meta_score.append(meta.text)
    else:
      meta_score.append(np.nan)

    # genre
    if gen is not None:
      genre.append(gen.text)
    else:
      genre.append(np.nan)

    # total duration
    if duration is not None:
      dur.append(duration.text)
    else:
      dur.append(np.nan)

  # seperating directors and stars list
  new = (pd.DataFrame(directors_and_stars))[0].str.split("Stars:", n = 1, expand = True)

  for i in new[0]:
    if i is not None:
      dir.append(i)
    else:
      dir.append(None)

  for i in new[1]:
    if i is not None:
      stars.append(i)
    else:
      stars.append(None)

  complete=[list(x) for x in zip(movie_name,rating,desc,votes,dir,stars,meta_score,genre,dur)]
  return complete
```

```python
year=2020
number_of_movies=1000
url1='https://www.imdb.com/search/title/?release_date=2020&sort=num_votes,desc&page=1'
result=[]

if number_of_movies==50:
  result.append(IMDB_scraping(year,number_of_movies,url1))
else:
  result.append(IMDB_scraping(year,number_of_movies,url1))
  i=51
  while(i != number_of_movies+1):
    result.append(IMDB_scraping(year,number_of_movies,'https://www.imdb.com/search/title/?release_date=2020-01-01,2020-12-31&sort=num_votes,desc&start='+str(i)+'&ref_=adv_nxt'))
    i+=50
result
```

```python
# changing list of list of list to list of list
final=[]
for sublist1 in result:
  for sublist2 in sublist1:
    final.append(sublist2)
final
```

```python
import pandas as pd
df=pd.DataFrame(final,columns=['Movie Name','IMDB Rating','Description','Votes','Directors','Stars','Metascore','Movie Genre','Total Duration'])
df=df.replace('\n','',regex=True)
df=df.replace('Directors:','',regex=True)
df=df.replace('Director:','',regex=True)
df=df.replace('\|','',regex=True)
df
```
